```python
import numpy as np
import datetime
growth = []
i = 2020
date = []
values = []
value = df['Levered Free Cash Flow'].tail(1)[0]
while i < 2025:
    date_year = datetime.date(i,1,1)
    date.append(date_year.year)
    value = value*(1+df['Growth Rate UnCFC'].tail(1)[0])
    values.append(value)
    print(value)
    i = i +1
growth.append(df['Growth Rate UnCFC'].tail(1)[0])
df_FCF = pd.DataFrame((date),columns=['date'])
df_FCF['simFCFLY'] = values
df_FCF.head()
```
